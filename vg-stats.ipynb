```python
import pandas as pd
import numpy as np
```

```python
games = pd.read_csv('./vgsales.csv')
games.head()
```

```python
#Which company is the most common game publisher?
pd.DataFrame(games['Publisher'].groupby(games.Publisher).count().sort_values(ascending=False)).head(1)
```

Electronic Arts is the most common Publisher

# What is the most common platform

```python
pd.DataFrame(games['Platform'].groupby(games.Platform).count().sort_values(ascending=False)).head(1)
```

DS is the most common platform

# What is the most common genre?

```python
pd.DataFrame(games['Genre'].groupby(games.Genre).count().sort_values(ascending=False)).head(1)
```

Action is the most common genre.
